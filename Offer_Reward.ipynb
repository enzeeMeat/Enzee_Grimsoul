```python
import cv2 
import matplotlib.pyplot as plt
import numpy as np



# test image 
image = cv2.imread('offer_capture.png') 
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY) 
histogram = cv2.calcHist([gray_image], [0], 
None, [256], [0, 256]) 

# data1 image 
image = cv2.imread('Template_Reward.png') 
gray_image1 = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY) 
histogram1 = cv2.calcHist([gray_image1], [0], 
None, [256], [0, 256]) 

# data2 image 
image = cv2.imread('Template_Offer.png') 
gray_image2 = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY) 
histogram2 = cv2.calcHist([gray_image2], [0], 
None, [256], [0, 256]) 


c1, c2 = 0, 0

# Euclidean Distace between data1 and test 
i = 0
while i<len(histogram) and i<len(histogram1): 
    c1+=(histogram[i]-histogram1[i])**2
    i+= 1
    c1 = c1**(1 / 2) 


# Euclidean Distace between data2 and test 
i = 0
while i<len(histogram) and i<len(histogram2): 
    c2+=(histogram[i]-histogram2[i])**2
    i+= 1
    c2 = c2**(1 / 2) 

if(c1<c2):
    fig = plt.figure("Reward!")
    plt.suptitle("Reward is more similar to Test_image as compare to Offer") 
    # show first image
    ax = fig.add_subplot(1, 2, 1)
    plt.imshow(cv2.imread('offer_capture.png') , cmap = plt.cm.gray)
    plt.axis("off")
    # show the second image
    ax = fig.add_subplot(1, 2, 2)
    plt.imshow(cv2.imread('Template_Reward.png'), cmap = plt.cm.gray)
    plt.axis("off")
    # show the images
    plt.show()

else:
    fig = plt.figure("Offer")
    plt.suptitle("Template_Offer is more similar to  Test_image as compare to Reward") 
    # show first image
    ax = fig.add_subplot(1, 2, 1)
    plt.imshow(cv2.imread('offer_capture.png') , cmap = plt.cm.gray)
    plt.axis("off")
    # show the second image
    ax = fig.add_subplot(1, 2, 2)
    plt.imshow(cv2.imread('Template_Offer.png') , cmap = plt.cm.gray)
    plt.axis("off")
    # show the images
    plt.show()

```
